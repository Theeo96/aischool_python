```python
a = 1234
b = 1

c = (a % (10 ** b)) // (10 ** (b - 1))

print(f'숫자 {a}의 {b}번째 자리의 수는 {c}')

print('\n각 자릿수 전체출력\n')

for b in range(1,5) :
    c = (a % (10 ** b)) // (10 ** (b - 1))
    print(f'숫자 {a}의 {10 ** (b - 1)}의자리 수는 {c}')
```

### 지민이는 온라인 게임을 하던 도중 갑작스럽게 적을 만났습니다. 지민이의 캐릭터는 마나를 소모하며 공격하는 캐릭터이고 한 번 공격 시 마나 10을 소모하며 대미지는 20입니다.

### 지민이의 닉네임 my_nic과 적의 닉네임 enemy_nic이 문자열로 주어지고, 지민이의 마나량 mana, 적의 체력 hp가 정수로 주어질 때 지민이가 남은 마나로 적을 쓰러뜨릴 수 있다면 "my_nic kill enemy_nic"을 쓰러뜨릴 수 없다면 "enemy_nic kill my_nic"을 출력하도록 코드를 작성해보세요. 

#### 제한사항 

##### 머쓱이의 캐릭터는 마나가 10보다 적으면 공격할 수 없습니다.<br>1 ≤ my_nic의 길이, enemy_nic의 길이 ≤ 10<br>1 ≤ mana, hp ≤ 1,000

 
##### 입출력 예 

##### 입력 #1 
##### abc<br>cba<br>100<br>200

##### 출력 #1 
##### abc kill cba 


##### 입력 #2 
##### jimin<br>enemy<br>200<br>1000 

##### 출력 #2 
##### enemy kill jimin 
<br>

##### 입출력 예 설명 

##### 입출력 예 #1 
##### 마나 100으로 지민이가 입힐 수 있는 대미지는 총 200입니다. 적의 hp가 200이므로 적을 쓰러뜨릴 수 있습니다. 따라서 "abc kill cba"를 출력해야 합니다. 


##### 입출력 예 #2 
##### 마나 200으로 지민이가 입힐 수 있는 대미지는 총 400입니다. 적의 hp가 1000이므로 적을 쓰러뜨릴 수 없습니다. 따라서 "enemy kill jimin"을 출력해야 합니다. 


```python
my_nic = input("지민이의 닉네임을 입력하세요 : ")
enemy_nic = input("적의 닉네임을 입력하세요 : ")
mana = int(input("지민이의 마나량(mana)을 입력하세요 : "))
hp = int(input("적의 체력(hp)을 입력하세요 : "))

if mana*2 > hp :
    print(f'{my_nic} kill {enemy_nic}')
else :
    print(f'{enemy_nic} kill {my_nic}')
```

### *을 가로로 a개, 세로로 b개 사용하여 직사각형을 만들려고 합니다. 정수 a, b가 매개변수로 주어질 때, a x b 직사각형을 출력하도록 코드를 작성하세요 

```python
def make_square(width, height) :

    for i in range(height) :
        print("*"*width)

# 넘어오는 매개변수가 없으므로 입력으로 구현
a, b = map(int, input("가로와 세로의 개수를 입력하세요 : ").strip().split())

make_square(a, b)
```

## -- 업그레이드 미션 12 --

### 과일가게를 운영해 주세요 
### f = ['사과', '바나나', '오렌지', '포도', '수박'] 
### c = [3, 3, 3, 3, 3] 



```python
fruit_list = ['사과', '바나나', '오렌지', '포도', '수박']
fruit_stock = [3, 3, 3, 3, 3]
command = ""
index = -1

print('과일 가게에 오신 것을 환영합니다!')
print(f'오늘 판매하는 과일 : {list(fruit_list)}')

while command != '그만' :

    if sum(fruit_stock) <= 0 :
        print('\n모든 과일이 다 팔렸습니다. 가게를 종료합니다.')
        break

    print(f'\n현재 재고 : {list(fruit_stock)}')

    #Jupyter 환경에서는 input message 가 안떠서, 가독성을 위해 출력
    print('어떤 과일을 구매하시겠습니까? (종료하려면 \'그만\'을 입력하세요) : ', end='')
    command = input('어떤 과일을 구매하시겠습니까? (종료하려면 \'그만\'을 입력하세요) : ')
    print(command)

    if command in fruit_list :
        index = fruit_list.index(command)
        
        if fruit_stock[index] <= 0 :
            print('해당 과일은 매진되었습니다! 다른 과일을 선택해주세요')
            continue

        fruit_stock[index] -= 1
        print(f'{fruit_list[index]}을(를) 구매하셨습니다! 남은 {fruit_list[index]}의 재고 : {fruit_stock[index]}')
        continue
    
    else :
        print('해당 과일은 존재하지 않습니다! 다시 입력해주세요')
        continue

```
