## 업그레이드미션 2  

### Todo: 변수 a에 저장된 값의 b자리 값을 계산하여	 변수 c에 저장하는 코드를 설계하라  

##### 예를 들어 a에 1234가 저장되어 있고 b가 1일 경우 첫 번째 자리값인 4가 c에 저장된다.  

##### a가 1234일때 b가 4일 경우는 4번째 자리값인 1이 c에 저장되어야 한다.  

```python
a = 1234
b = 1

c = (a % (10 ** b)) // (10 ** (b - 1))

print(f'숫자 {a}의 {b}번째 자리의 수는 {c}')

print('\n각 자릿수 전체출력\n')

for b in range(1,5) :
    c = (a % (10 ** b)) // (10 ** (b - 1))
    print(f'숫자 {a}의 {10 ** (b - 1)}의자리 수는 {c}')
```

## 업그레이드미션 3 

### 삼각형의 넓이 구하기 

##### 사용자에게 삼각형의 밑변과 높이를 입력 받아  

##### 삼각형의 넓이를 출력하는 코드를 작성하세요. 

```python
width = int(input("삼각형의 밑변 길이를 입력하세요 : "))
height = int(input("삼각형의 높이를 입력하세요 : "))
area = (width * height) / 2

print(f'삼각형의 밑변 길이는 {width}, 높이는 {height}이며')
print(f'삼각형의 넓이는 {area} 입니다')
```
